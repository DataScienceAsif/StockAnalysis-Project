```python
# Question 1: Use yfinance to Extract Stock Data
# Reset the index, save, and display the first five rows of the tesla_data dataframe using the head function.
# Upload a screenshot of the results and code from the beginning of Question 1 to the results below.

# Import required libraries
import yfinance as yf
import pandas as pd

# Fetch Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Reset the index
tesla_data.reset_index(inplace=True)

# Save the data to a CSV file (optional)
tesla_data.to_csv("tesla_data.csv", index=False)

# Display the first five rows of the dataframe
print(tesla_data.head())
```

```python
# Question 2: Use Webscraping to Extract Tesla Revenue Data
# Display the last five rows of the tesla_revenue dataframe using the tail function. Upload a screenshot of the results.

import yfinance as yf
import pandas as pd
# Fetch Tesla stock data using yfinance
tesla = yf.Ticker("TSLA")

# Get the financials (income statement) for Tesla
tesla_financials = tesla.financials
# Extract the revenue row (this row usually contains the revenue data)
tesla_revenue = tesla_financials.loc['Total Revenue']

# Convert Series to DataFrame and transpose for better readability
tesla_revenue = tesla_revenue.to_frame()
# Reset the index and rename the columns (we need to adjust this)
tesla_revenue = tesla_revenue.reset_index()

# Rename the columns for clarity
tesla_revenue.columns = ['Date', 'Revenue']
# Convert revenue to a float and date to datetime
tesla_revenue['Revenue'] = tesla_revenue['Revenue'].astype(float)
tesla_revenue['Date'] = pd.to_datetime(tesla_revenue['Date'])

# Display the last 5 rows of the revenue data
print(tesla_revenue.tail())

```

```python
# Question 4: Use Webscraping to Extract GME Revenue Data
# Display the last five rows of the gme_revenue dataframe using the tail function. Upload a screenshot of the results.

import yfinance as yf
import pandas as pd

# Get the GameStop (GME) data
gme = yf.Ticker("GME")

# Fetch the financials (including revenue) for GME
financials = gme.financials

# Extract the row containing 'Total Revenue'
revenue_data = financials.loc['Total Revenue']

# Convert to DataFrame for better readability
gme_revenue = pd.DataFrame(revenue_data).transpose()

# Display the last 5 rows
print(gme_revenue.tail())
```

```python
# Question 5: Plot Tesla Stock Graph
# Use the make_graph function to graph the Tesla Stock Data, also provide a title for the graph.

import yfinance as yf
import matplotlib.pyplot as plt

# Fetch Tesla stock data
tesla = yf.Ticker("TSLA")
# Get the historical stock data (adjust the period as needed)
tesla_data = tesla.history(period="1y")  # For 1 year of data
# Define the make_graph function
def make_graph(data, title="Tesla Stock Price"):
    # Plot the closing price of Tesla stock
    plt.figure(figsize=(10,6))
    plt.plot(data['Close'], label='Tesla Stock Price', color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()
# Call the make_graph function to plot the graph
make_graph(tesla_data)
```
